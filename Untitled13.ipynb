## 회전 좌표값 찾기 

```python
img=cv2.imread('Downloads/Kitti_traning_dataset_5/000000.png')
df=pd.read_csv('Downloads/Kitti_traning_dataset_5/000000.txt',sep=' ',header=None)
```

```python
def distance(x1, y1, x2, y2):
    result = math.sqrt( math.pow(x1 - x2, 2) + math.pow(y1 - y2, 2))
    return result
```

```python
def boundingbox(img,df):
    for X in range (0,len(df)):
        label=df[0][X] 
        x1=round(float(df[4][X])) 
        y1=round(float(df[5][X]))
        x2=round(float(df[6][X]))
        y2=round(float(df[7][X]))
        h, w, c = img.shape
        print(h/2,w/2)
        img=cv2.rectangle(img,(x1,y1),(x2,y2),(0,255,0),1)
        img=cv2.putText(img, label, (x1-10,y1-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,255,0), 1)
        #img=cv2.circle(img,(round(img.shape[1]/2),round(float(img.shape[0]/2)),1,(0,255,0),-1))
        img=cv2.circle(img, (round(w/2), round(h/2)), 10, (255, 0, 0), -1)
        img=cv2.line(img,(round(w/2), round(h/2)),(x1,y1),(255, 0, 0),5)
        img=cv2.line(img,(round(w/2), round(h/2)),(x2,y2),(255, 0, 0),5)
        strr1= '('+str(x1)+','+str(y1)+')'
        strr2= '('+str(x2)+','+str(y2)+')'
        strr4= '('+str(round(w/2))+','+str(round(h/2))+')'
        img=cv2.putText(img,strr1, (x1,y1), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,255), 1)
        img=cv2.putText(img,strr2, (x2,y2), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,255), 1)
        img=cv2.putText(img,strr4, (round(w/2), round(h/2)), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,255), 1)
        d1 = distance(x1, y1, round(w/2), round(h/2))
        print('중심점에서 첫번째 좌표까지 거리:',d1)
        d2 = distance(x2, y2, round(w/2), round(h/2))
        print('중심점에서 두번째 좌표까지 거리:',d2)

        
    plt.imshow(img)
    plt.show()
    cv2.imshow('img',img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
def 회전좌표값(x1,y1):    
    각도=60
    라디안 = 각도 * (float)(math.pi / 180)
    x2 = round(math.cos(라디안) * x1 - math.sin(라디안) * y1)
    y2 = round(math.sin(라디안) * x1 + math.cos(라디안) * y1)
    return x2,y2
```
